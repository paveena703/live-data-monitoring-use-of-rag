```python
import yfinance as yf
import smtplib
import time
from email.mime.text import MIMEText
from datetime import datetime

# ===== EMAIL CONFIG =====
SENDER_EMAIL = "paveenaparanthaman@gmail.com"
APP_PASSWORD = "pdhh ntop nfmw kxsi"
RECEIVER_EMAIL = "paveenaparanthaman@gmail.com"

# ===== STOCK =====
STOCK_SYMBOL = "RELIANCE.NS"
STOCK_NAME = "RELIANCE"

previous_price = None

while True:
    try:
        stock = yf.Ticker(STOCK_SYMBOL)
        current_price = stock.fast_info["last_price"]

        now = datetime.now().strftime("%d-%m-%Y %H:%M:%S")

        # Price comparison
        if previous_price is None:
            status = "‚è≥ First data received"
            change = 0
        else:
            change = round(current_price - previous_price, 2)
            if change > 0:
                status = "üìà PRICE INCREASED"
            elif change < 0:
                status = "üìâ PRICE DECREASED"
            else:
                status = "‚ûñ NO CHANGE"

        body = f"""
üìä LIVE STOCK UPDATE

Stock Name : {STOCK_NAME}
Price      : ‚Çπ{current_price}
Change     : {change}
Status     : {status}
Time       : {now}

(Update every 30 seconds)
"""

        msg = MIMEText(body)
        msg["From"] = SENDER_EMAIL
        msg["To"] = RECEIVER_EMAIL
        msg["Subject"] = f"üìà {STOCK_NAME} Live Price Update"

        server = smtplib.SMTP("smtp.gmail.com", 587)
        server.starttls()
        server.login(SENDER_EMAIL, APP_PASSWORD)
        server.send_message(msg)
        server.quit()

        print("‚úÖ Mail sent |", status, "| Price:", current_price)

        previous_price = current_price

    except Exception as e:
        print("‚ùå Error:", e)

    time.sleep(30)  # ‚è±Ô∏è 300 seconds interval
```
